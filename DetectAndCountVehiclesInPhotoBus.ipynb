In this notebook I am using haar cascade to detect buses in an image, here I would be using the pre-trained XML file already uploaded in google drive

```python
# Import Libs
import pandas as pd
import numpy as np
import cv2
from google.colab import drive
from PIL import Image
```

```python
# Locate search the image of bus which I have already uploaded on google drive
!ls '/content/drive/MyDrive/DataSets/DetectAndCountVehicles/Required Files/'
#!ls '/content/drive/MyDrive/DataSets/DetectAndCountVehicles/Data'
```

```python
# read the image 
bus_image = Image.open('/content/drive/MyDrive/DataSets/DetectAndCountVehicles/Data/bus.jpg')
bus_image
```

```python
# Resize
bus_image= bus_image.resize((450,250))

# Convert into arraya
bus_image_array = np.array(bus_image)

# Convert into gray scale
bus_gray = cv2.cvtColor(bus_image_array, cv2.COLOR_BGR2GRAY)

Image.fromarray(bus_gray) # View the gray scale Image
```

```python
# Load the haar cascade
bus_cascade_src = '/content/drive/MyDrive/DataSets/DetectAndCountVehicles/Required Files/Bus_front.xml'
bus_cascade = cv2.CascadeClassifier(bus_cascade_src)

# Run
bus = bus_cascade.detectMultiScale(bus_gray, 1.1,1)
```

```python
# running the loop & counting the buses in the image
cnt = 0
for (x,y,w,h) in bus:
  cv2.rectangle(bus_image_array, (x,y), (x+w, y+h), (0,255,0),2)
  cnt+=1
print(cnt, 'Buses Found')
Image.fromarray(bus_image_array)
```
